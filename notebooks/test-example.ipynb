```python
import numpy as np
import healpy as hp

from polymv import polymv
```

```python
help(polymv)
```

```python
lmax = 1000
```

```python
%%time
np.random.seed(1)
alm = hp.synalm(np.ones(2000), lmax=lmax)
alm = np.column_stack([alm.real, alm.imag])
alm
```

```python
%%time
mvs = polymv.mvs(alm[:, 0], alm[:, 1], lmax)
mvs
```

```python
%%time
fvs = polymv.fvs(mvs[0], mvs[1], lmax)
fvs
```
